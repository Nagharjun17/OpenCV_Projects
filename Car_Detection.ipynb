```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = cv2.imread('/content/rd1.jpg')
img1 = cv2.imread('/content/rd2.jpg')
img2 = cv2.imread('/content/r3.jpg')
```

```
def display(img):
    fig = plt.figure(figsize = (10, 8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```
display(img)
display(img1)
display(img2)
```

```
car_classifier = cv2.CascadeClassifier('/content/haarcascade_car.xml')
```

```
def detect_car(img):
  car_img = img.copy()
  car_rects = car_classifier.detectMultiScale(car_img, 1.1, 1)
  for(x, y, w, h) in car_rects:
    cv2.rectangle(car_img, (x, y), (x + y, y + h), (255, 255, 255), 5)
  return car_img
```

```
result = detect_car(img)
result1 = detect_car(img1)
result2 = detect_car(img2)
```

```
display(result)
display(result1)
display(result2)
```
